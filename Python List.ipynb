### Question 1: Write a Python program to count the number of elements in a list without using len.

### Question 2: Write a Python program to find the maximum and minimum values in a given list.

- `sample input : `
    - `var = [10,11,12,13,14,15]`
- `sample output :`
    - `result : max = 15 , min = 10`

```python
# Please write The solution Here for above question 2:
var=[10,11,12,13,14,15]
#print(max(var))
#print(min(var))
max_value=var[0]
min_value=var[0]
for i in range(1,len(var)):
  if max_value>var[i]:
    max_value=max_value
  else:
    max_value=var[i]
for j in range(len(var)):
  if min_value<var[j]:
    min_value=min_value
  else:
    min_value=var[j]
print(f"max = {max_value} , min = {min_value}")
```

### Question 3: Write a Python program to remove the first occurrence of a specified element from a list.

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : [1,3,4,2,5]`
    

```python
# Please write The solution Here for above question 3:
var=[1,2,3,4,2,5]
for i in var:
  x=var.count(i)
  if x>1:
    var.remove(i)
print(var)
```

### Question 6:  Write a Python program to find the index of a specified element in a list..

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 3 index is 2`

```python
# Please write The solution Here for above question 6:
var=[1,2,3,4,2,5]
n=int(input())
if n in var:
  for i in range(len(var)):
    if var[i]==n:
      print(f'{n} index is {i}')
else:
  print(f'{n} not in given list')
```

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
# Please write The solution Here for above question 7:
var=[134,121,155]
for i in var:
  s=""
  temp_variable=i
  while temp_variable>0:
    r=temp_variable%10
    temp_variable=temp_variable//10
    s=s+str(r)
  if i==int(s):
    print(f"{i} is a Palindrome")
  else:
    print(f'{i} is not a palindrome')
```

### Question 9: Write a Python program to remove the duplicates from a list

- `sample input : `
    - `var = [10,11,12,10,12,24]`
- `sample output :`
    - `result : [10,11,12,24]`

```python
# Please write The solution Here for above question 9:
var=[10,11,12,10,12,24]
for i in var:
  x=var.count(i)
  #print(x)
  if x>1:
    value=var.index(i)
    new_value=var.index(i,value+1)
    #print(new_value)
    var.pop(new_value)
print(var)

```

### Question 10:  Write a Python program to sort a list of strings in Ascii Order [Ascending order]


- `sample input : `
    - `var = ['a' , 'c' , 'b']`
- `sample output :`
    - `result : ['a' , 'b' , 'c']`


```python
# Please write The solution Here for above question 10:
var=['a','c','b']
list1=[]
new_list=[]
final_chr_list=[]
for i in var:
  list1.append(ord(i))
  for j in range(1,len(list1)):
    first_value=list1[0]
    for k in range(1,len(list1)):
      if first_value<list1[k]:
        first_value=first_value
      else:
        first_value=list1[k]
    list1.remove(first_value)
    new_list.append(first_value)
final_ord_list=new_list+list1
#print(final_ord_list)
#print(type(final_ord_list))
for character in final_ord_list:
  final_chr_list.append(chr(character))
print(final_chr_list)
```
